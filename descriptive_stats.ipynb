Q8. Outlier Detection using IQR

```python
upper_length=df['sepal_length'].quantile(0.75)+1.5*sepal_length_IQR
upper_length
```

```python
lower_length=df['sepal_length'].quantile(0.25)-1.5*sepal_length_IQR
lower_length
```

```python
outliers_length=df[(df['sepal_length']>upper_length) |
                  (df['sepal_length']>lower_length)]
print(outliers_length)
```
